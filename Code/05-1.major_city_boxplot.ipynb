# 5-1.서울, 경기, 대구, 경북 확진자수 boxplot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

import os
import datetime as dt
from pprint import pprint as pp 
from inspect import signature as sig
```

```python
# 변수명 case 로 case.csv 파일 불러옴 
case = pd.read_csv('./DataScienceforCOVID19/Case.csv')
```

```python
# 서울, 경기, 대구, 경북 확진자수

case_sgdg = case[case.province.isin
                 (['Seoul', 'Gyeonggi-do' , 'Daegu' , 'Gyeongsangbuk-do'])]

case_sgdg
```

```python
# 서울, 경기, 대구. 경북 지역의 케이스별 확진자수 Boxplot (6/30 기준)

frame = plt.figure(figsize = ( 8 , 10 ) )
grid = plt.xticks (rotation = 60)
grid = sns.boxplot(x = 'province' , y = 'confirmed' , width=0.5, data = case_sgdg, palette = sns.color_palette("deep"))
grid.set_title('Confirmed Cases in 4 Provinces as of June 30, 2020', fontsize = 20, pad = 10)
grid.set_xlabel('Province')
grid.set_ylabel('Number of Confirmed Patient')
```
